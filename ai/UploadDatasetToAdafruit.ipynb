```python
import numpy as np
import matplotlib.pyplot as plt
import datetime as dt
import time
import pandas as pd
from Adafruit_IO import Client
```

```python
feed = "location-1"
feed_data = pd.read_csv(feed + ".csv", sep=';')
feed_data
```

```python
# da impostare con i dati relativi ad account e feed
username = None
aio_key = None
feed_key = 'location-1'
samples_sent = 0

aio = Client(username, aio_key)
metadata = { 
    "lat": None,
    "lon": None,
    "ele": None,
    "created_at": None
}

# upload dei campioni su feed Adafruit IO a intervalli 
# di 2 secondi (=> con frequenza di 30 datapoint/min)
for i in range(feed_data.shape[0]):
  metadata["created_at"] = feed_data.loc[i, "created_at"]
  aio.send_data(feed_key, int(feed_data.loc[i, "value"]), metadata=metadata)
  samples_sent += 1
  time.sleep(2)

print(f"Sent { samples_sent } samples to feed Adafruit IO.")
```
